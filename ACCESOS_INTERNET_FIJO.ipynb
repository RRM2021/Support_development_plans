```python
# Convert the values in the column to numeric data type
df['ID_MUNICIPIO'] = pd.to_numeric(df['ID_MUNICIPIO'], errors='coerce')

```

```python
# Create DataFrame
new_df2 = spark.createDataFrame(df)

# Show DataFrame
new_df2.show()
```

```python
# Register DataFrame as a temporary table
new_df2.createOrReplaceTempView("N")
```

```python
# Register the DataFrame as a temporary view
new_df2.createOrReplaceTempView("temp_view")

# Drop the column using SQL and create a new DataFrame without the dropped column
new_df3 = spark.sql("SELECT ANNO, TRIMESTRE, EMPRESA, ID_MUNICIPIO, MUNICIPIO, ID_DEPARTAMENTO, DEPARTAMENTO, ID_SEGMENTO, SEGMENTO, VELOCIDAD_EFECTIVA_DOWNSTREAM, VELOCIDAD_EFECTIVA_DOWNSTREAM, ID_TECNOLOGIA, TECNOLOGIA, ACCESOS FROM temp_view")

# Show the new DataFrame
new_df3.show()
```

```python
# Register the DataFrame as a temporary view
#df.createOrReplaceTempView("temp_view")

# Use SQL query to select rows with null values based on a specific column
result = spark.sql("SELECT * FROM temp_view WHERE ID_MUNICIPIO IS NULL")
result.show()
```

```python
# Iterate over each column in the DataFrame
for col in new_df3.columns:
    # Use SQL query to select rows with null values for the current column
    result = spark.sql(f"SELECT * FROM temp_view WHERE {col} IS NULL")

    # Display the column name and the rows with null values for that column
    print(f"Column: {col}")
    result.show()
```

```python
# Perform SQL query to calculate the mean of VELOCIDAD_EFECTIVA_DOWNSTREAM for each MUNICIPIO and order by VELOCIDAD_EFECTIVA_DOWNSTREAM
result_df = spark.sql("SELECT MUNICIPIO, AVG(VELOCIDAD_EFECTIVA_DOWNSTREAM) AS mean_VELOCIDAD_EFECTIVA_DOWNSTREAM FROM temp_view GROUP BY MUNICIPIO")

# Show the result DataFrame
# Display the last 5 rows of the DataFrame
result_df.show()
```

```python
# Register the DataFrame as a temporary view
result_df.createOrReplaceTempView("result_df")

# Select rows where MUNICIPIO is either "GUARNE" or "MEDELLÍN" using SQL
selected_df = spark.sql("SELECT * FROM result_df WHERE MUNICIPIO IN ('GUARNE', 'MEDELLÍN', 'BOGOTÁ', 'RIONEGRO', 'MARINILLA', 'CALI' 'EL RETIRO')")

# Show the resulting DataFrame
selected_df.show(10)

```

```python
# Convert Spark DataFrame to Pandas DataFrame
pandas_df = result_df.toPandas()
pandas_df
```

```python
# Set display options to show all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Order the DataFrame based on the mean_VELOCIDAD column
ordered_df = pandas_df.sort_values(by='mean_VELOCIDAD_EFECTIVA_DOWNSTREAM')
# Display the ordered DataFrame
ordered_df
```
