```python
#pd.options.mode.copy_on_write = False

df_cl1 = df_cl.copy()  # Explicitly create a copy


# Iterate over each column in the DataFrame
for column in df_cl.columns:
    # Check if the column data type is object
    if df_cl1[column].dtype == 'object':
        # Calculate the mode of the column
        mode_value = df_cl1[column].mode()[0]
        # Replace null values with the mode
        df_cl1[column].fillna(mode_value, inplace=True)

# Print the DataFrame after replacing null values in object columns
df_cl1.info()
print(df_cl1.shape)
```

```python
import pandas as pd
df_cl2 = df_cl1.copy()  # Explicitly create a copy

# Dictionaries to store the unique categories for each column
unique_categories = {}

# Iterate over columns with object data type
for column in df_cl2.select_dtypes(include=['object']).columns:
    # Convert categorical unique values into numerical values
    df_cl2[column], uniques = pd.factorize(df_cl2[column])
    unique_categories[column] = uniques

print("\nDataFrame after converting categorical values to numerical values:")
df_cl2

```

```python
# Function to map numerical values back to categorical values
def map_back(df, unique_categories):
    for column, uniques in unique_categories.items():
        df[column] = df[column].map(lambda x: uniques[x])
    return df

# Map back the numerical values to the original categorical values
df_original = map_back(df.copy(), unique_categories)

print("\nDataFrame after mapping numerical values back to categorical values:")
print(df_original)
```

```python
# Surpress warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np

# Load data (digits dataset)
data = Clus_dataSet
pca = PCA(3)
df_dim_red = pca.fit_transform(data)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
from sklearn.model_selection import GridSearchCV
```

```python
k_means = KMeans()
# Set up the parameter grid to search over
param_grid = {"n_clusters": [2, 3, 4, 5], "n_init": [4, 5], "max_iter": [5, 10, 15, 25], "algorithm": ["auto", "full", "elkan"]}

# Set up the grid search
grid_search = GridSearchCV(k_means, param_grid=param_grid, cv=5)

# Run the grid search
grid_search.fit(df_dim_red)

# Print the best parameters and the corresponding score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
from sklearn.cluster import KMeans

# Define the best parameters obtained from the grid search
best_params = grid_search.best_params_
#best_params = {'algorithm': 'full', 'max_iter': 5, 'n_clusters': 3, 'n_init': 5}
# Create a K-Means model with the best parameters
k_means = KMeans(**best_params)

# Fit the model to your data
k_means.fit(df_dim_red)

# Get the cluster labels for each data point
labels = k_means.labels_
print(labels)
```

CLUSTERS VISUALIZATION

```python
# Plot individual clusters

fig = plt.figure(figsize=(10, 8))  # Adjust the figsize as needed
ax = fig.add_subplot(111, projection='3d')

for k in range(len(k_means.cluster_centers_)):
    filtered_data = df_dim_red[labels == k]
    ax.scatter(filtered_data[:, 0], filtered_data[:, 1], filtered_data[:, 2], label=f'Cluster {k}')

ax.set_title('K-Means Clusters')
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get the unique groups in 'Clus_km'
groups = df_cl1['Clus_km'].unique()

# Set up the figure for the subplots
fig, axes = plt.subplots(len(groups), 1, figsize=(10, 6 * len(groups)))

# Iterate over each group
for idx, group in enumerate(groups):
    # Filter the dataframe for the current group
    group_df = df_cl1[df_cl1['Clus_km'] == group]

    # Initialize a dictionary to store percentage data for each column
    percentage_data = {}

    # Iterate over each object column
    for col in df_cl1.select_dtypes(include=['object']).columns:
        if col != 'Clus_km':
            # Get the value counts for the column in the current group
            value_counts = group_df[col].value_counts(normalize=True) * 100
            percentage_data[col] = value_counts

    # Create a DataFrame from the percentage data
    percentage_df = pd.DataFrame(percentage_data).fillna(0)

    # Transpose the DataFrame to get the columns as stacked bars
    percentage_df = percentage_df.T

    # Plot the percentage DataFrame as a stacked bar plot
    percentage_df.plot(kind='bar', stacked=True, ax=axes[idx])
    axes[idx].set_xlabel('Features')
    axes[idx].set_ylabel('Percentage')
    axes[idx].set_title(f'Percentage of unique values for group {group}')
    axes[idx].legend(title='Unique Values', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', markerscale=0.8)

plt.tight_layout()
plt.show()
```
